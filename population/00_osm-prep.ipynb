```python
import pyrosm
from os import path
import time
```

```python
# constants
PBF_PATH = path.join(
    '../data/raw/osm', pyrosm.data.sources.africa.egypt['name'])
    
BOUNDING_BOX_GCR = [30.5, 29.5, 32.0, 30.5]  # [x_min, y_min, x_max,y_max]
```

```python
# Download Geofabrik PBF for Egypt
if not path.exists(PBF_PATH):
    print(f"Downloading {pyrosm.data.sources.africa.egypt['name']} to {PBF_PATH}")
    pyrosm.get_data(dataset='egypt', directory='../data/raw/osm/')
else:
    print('PBF file exists. Download skipped.')
```

```python
# Initialize PYROSM reader
osm_reader = pyrosm.OSM(filepath=PBF_PATH, bounding_box=BOUNDING_BOX_GCR)
```

```python
# Extracting Road Network

ROAD_NETWORK_FILE_PATH = path.join(
    '../data/interim', 'osm', 'osm_road_network_gcr.gpkg')

if not path.exists(ROAD_NETWORK_FILE_PATH):
    # Parse road network from PBF
    print('Extracting road network from PBF file')
    start_time = time.time()
    road_network = osm_reader.get_network(network_type='driving')
    print(f'Parsing road network successful. Time elapsed: {time.time() - start_time}')

    # Save road network to disk
    print(f'Saving road network GeoDataFrame to {ROAD_NETWORK_FILE_PATH}')
    start_time = time.time()
    road_network.to_file('../data/interim/osm/osm_road_network_gcr.gpkg')
    print(f'Saving road network successful. Time elapsed: {time.time() - start_time}')

else:
    print('Parsing and saving road network skipped.')
```

```python
# Extracting Buildings
buildings = osm_reader.get_buildings()
```
